```python
class preprocess():
    def read_excel(path):
        data = pd.read_excel(path)
        return data
        
    def nlp(data):
        non_space_sentense = []
        for dat in data['단어']:
            non_space_sentense.append(str(dat).replace(' ', '')) ## 스페이스 제거
            
        non_space_sentense = pd.DataFrame(non_space_sentense)
        add_df = pd.merge(data, non_space_sentense, how = 'outer', left_index=True, right_index=True)
        add_df = add_df.rename({0:'non_space'}, axis='columns')  ## 스페이스 제거된 열 추가
        
        return add_df
    
    def over_lab_data(df):
        n = 0
        over_lab_data = pd.DataFrame()
        while n < len(df)-1:
            for i in range(n, n+1, 1):
                if df['non_space'].iloc[n] == df['non_space'].iloc[n+1]:
                    over_lab_data = over_lab_data.append(df.iloc[n+1])
                else:
                    pass

            n += 1

        return over_lab_data
    
    def overlab_df_del(df):
        df = df.drop_duplicates(['non_space'])
        
        return df



        
```

```python
path = 'data/synonym_docs.xlsx'
data = preprocess.read_excel(path)
add_df = preprocess.nlp(data)
```
